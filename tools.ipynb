# Analyzation

## Groupby

```python
#Groupby and retrieve one group
column = 'Class 1'
group_key = 'Dark'
groups = df.groupby(column)
group = groups.get_group(group_key)
group
```

```python
#Groupby and count size of each group for each group
column = 'Class 1'
groups = df.groupby(column).count()
groups
```

```python
#Groupby with several groups
column = 'Class 1'
column2 = 'Class 2'
groups = df.groupby([column, column2]).count()
groups
```

# Visualization

```python
# Basic scatterplot with notation
sns.set()

column = 'HP'
column2 = 'Attack'
data = df[[column, column2]]

fig, ax = plt.subplots(figsize = (15, 10))
sns.scatterplot(x = column, y = column2, data = data, ax = ax)
ax.set_xlabel('XLABEL')
ax.set_ylabel('YLABEL')
ax.set_title('TITLE')
plt.show()
```

```python
# Matrix fig
fig, axs = plt.subplots(nrows = 3, ncols= 2, figsize = (15, 10))
for axx in axs:
    for axy in axx:
        data.plot(ax=axy)
        
# Vector fig
fig, axs = plt.subplots(nrows = 3, figsize = (15, 10))
for ax in axs:
      sns.scatterplot(x = column, y = column2, data = data, ax = ax)
```

```python
#Controlling colors
pal = sns.color_palette(n_colors=10)
sns.palplot(pal)

pal = sns.hls_palette(10)
sns.palplot(pal)

pal = sns.choose_dark_palette()

pal = sns.choose_cubehelix_palette()

pal = sns.choose_diverging_palette()
```
